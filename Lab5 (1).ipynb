```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

#ruta = '/content/drive/MyDrive/Colab_Notebooks/Miniproyecto1/'

ruta = '/content/drive/MyDrive/Se√±ales3/ArchivosMiniProy1/'
fileName = 'DataMP1.csv'
db = pd.read_csv(ruta+fileName)

labels = db.iloc[:,43]
db = db.iloc[:,1:43]
```

```python
scaler = StandardScaler()
scaler.fit(db)
print('var: ',scaler.var_)
print('mean: ',scaler.mean_)
```

```python
db_scaled = scaler.transform(db)
```

```python
acc_train = []
acc_test = []
Nrange = 100

for it in Nrange:
  x_train, x_test, y_train, y_test = train_test_split(db_scaled,labels,test_size=0.2)

  modelRL = LogisticRegression(C = 1.0)
  modelRL.fit(x_train, y_train)

  y_est_train = modelRL.predict(x_train)
  y_est_test = modelRL.predict(y_train)

  acc_train.append(accuracy_score(y_train,y_est_train))
  acc_test.append(accuracy_score(y_test,y_est_test))

print("ACC Train: ", acc_train.mean_, "+/-", acc_train.var_)
print("ACC Test: ", acc_test.mean_, "+/-", acc_test.var_)

```
